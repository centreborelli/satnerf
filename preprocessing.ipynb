```python
import glob
import os
import shutil
```

```python
base_dir = "/home/myid/zis35724/jupyter/DFC2019/Track3-RGB-1"
target_dir = "/home/myid/zis35724/jupyter/DFC2019/Track3-RGB-1-crops"
```

```python
paths = sorted(glob.glob("/home/myid/zis35724/jupyter/DFC2019/Track3-RGB-1/JAX_*"))
```

```python
aoi_ids = set(["JAX_" + p.split("/")[-1].split("_")[1] for p in paths])
```

```python
aoi_ids
```

```python
os.makedirs(target_dir, exist_ok=True)
```

```python
for id in aoi_ids:
    print("creating", target_dir + "/" + id)
    os.makedirs(target_dir + "/" + id, exist_ok=True)

    fs = sorted(glob.glob("/home/myid/zis35724/jupyter/DFC2019/Track3-RGB-1/" + id + "*"))
    for f in fs:
        print("copying", f)
        shutil.copy2(f, target_dir + "/" + id)
```

```python
from create_satellite_dataset import create_satellite_dataset
dfc_dir = "datasets/DFC2019"

for aoi_id in aoi_ids:
    print("Creating raw metadata for", aoi_id)
    output_dir = "datasets/Track3-preprocess/" + aoi_id + "/raw"
    os.makedirs(output_dir, exist_ok=True)
    create_satellite_dataset(aoi_id, dfc_dir, output_dir, ba=False, crop_aoi=False, splits=True)

    print("Creating ba metadata crop for", aoi_id)
    output_dir = "datasets/Track3-preprocess/" + aoi_id + "/ba"
    os.makedirs(output_dir, exist_ok=True)
    create_satellite_dataset(aoi_id, dfc_dir, output_dir, ba=True, crop_aoi=True, splits=True)
```
